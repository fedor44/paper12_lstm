```python
import pandas as pd

# Yahoo Finance에서 제공 금융 데이터 로드
import yfinance as yf

# 데이터 출처 Yahoo Finance
# 금융 데이터 로드
tickers = ['AAPL','MSFT','AVGO']
start = '2018-01-01'
end = '2023-05-31'
df_ticker_info = pd.DataFrame()
for ticker in tickers:
    ticker_yf = yf.Ticker(ticker)
    if df_ticker_info.empty:
        df_ticker_info = ticker_yf.history(start = start, end = end)
        df_ticker_info['ticker']= ticker
    else:
        data_temp = ticker_yf.history(start = start, end = end)
        data_temp['ticker']= ticker
        df_ticker_info = pd.concat([df_ticker_info, data_temp])

df_ticker_info
```

```python
import numpy as np

# pct_change 함수를 이용, 각 날짜별 종가를 이용하여 수익률을 계산
df_ticker_info['currRet'] = df_ticker_info['Close'].pct_change()

# ticker가 바뀔 때는 정확한 수익률을 계산할 수 없으므로 첫행은 NaN 처리
temp_df_ticker_info = df_ticker_info.loc[:, 'currRet'].copy()
for i in range(len(df_ticker_info)-1):
  if df_ticker_info.iloc[i]['ticker'] != df_ticker_info.iloc[i+1]['ticker']:
    temp_df_ticker_info.iloc[i+1] = np.nan

df_ticker_info['currRet'] = temp_df_ticker_info

# 전일, 당일, 후일의 데이터를 활용하여 이벤트 수익률을 계산하여 이벤트에 따른 주가의 변동을 보다 정확하게 측정
# currRet(t-1) + currRet(t) + currRet(t+1) -> 뉴스 데이터 전일, 당일, 후일의 수익
df_ticker_info['eventRet'] = df_ticker_info['currRet'] + df_ticker_info['currRet'].shift(-1) + df_ticker_info['currRet'].shift(1)

# 날짜 (년-월-일) 정리
df_ticker_info.reset_index(inplace=True)
df_ticker_info['date'] = pd.to_datetime(df_ticker_info['Date']).dt.date

df_ticker_info
```

```python
#LSTM start
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from google.colab import drive
drive.mount('/content/drive/')
import pandas as pd

# 주식 정보 및 감성 분석 데이터 파일
df_aapl = pd.read_csv('/content/drive/MyDrive/paper/stock/df_aapl.csv')
df_msft = pd.read_csv('/content/drive/MyDrive/paper/stock/df_msft.csv')
df_avgo = pd.read_csv('/content/drive/MyDrive/paper/stock/df_avgo.csv')

# separate dates for future plotting
#df_aapl_date = pd.to_datetime(df_aapl['date'])
#df_msft_date = pd.to_datetime(df_msft['date'])
#df_avgo_date = pd.to_datetime(df_avgo['date'])

# 필요한 데이터만 남기고 정리
df_aapl_1 = df_aapl[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'vaderSentiment']]
df_aapl_2 = df_aapl[['Open', 'High', 'Low', 'Close', 'Volume', 'finbertSentiment']]
df_aapl_3 = df_aapl[['Open', 'High', 'Low', 'Close', 'Volume', 'finetunedFinbertSentiment']]

# 특성 선택 및 데이터 정규화
features_1 = ['Open', 'High', 'Low', 'Close', 'Volume', 'vaderSentiment']
features_2 = ['Open', 'High', 'Low', 'Close', 'Volume', 'finbertSentiment']
features_3 = ['Open', 'High', 'Low', 'Close', 'Volume', 'finetunedFinbertSentiment']
# 결측치 중립으로 처리
df_aapl_1['vaderSentiment'] = df_aapl_1['vaderSentiment'].fillna(2.0)
df_aapl_2['finbertSentiment'] = df_aapl_2['finbertSentiment'].fillna(2.0)
df_aapl_3['finetunedFinbertSentiment'] = df_aapl_3['finetunedFinbertSentiment'].fillna(2.0)
```

```python
# import libraries
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import os

# 'save_weights' 디렉토리가 없다면 생성
if not os.path.exists('./save_weights'):
    os.makedirs('./save_weights')

stock_data = df_aapl_1

# save original 'Open' prices for later
original_open = stock_data['Open'].values

# separate dates for future plotting
dates = pd.to_datetime(stock_data['date'])

# variables for training
cols = list(stock_data)[1:6]

# new dataframe with only training data - 5 columns
stock_data = stock_data[cols].astype(float)

# normalize the dataset
scaler = StandardScaler()
scaler = scaler.fit(stock_data)
stock_data_scaled = scaler.transform(stock_data)

# split to train data and test data
n_train = int(0.9*stock_data_scaled.shape[0])
train_data_scaled = stock_data_scaled[0: n_train]
train_dates = dates[0: n_train]

test_data_scaled = stock_data_scaled[n_train:]
test_dates = dates[n_train:]

# data reformatting for LSTM
pred_days = 1  # prediction period
seq_len = 14   # sequence length = past days for future prediction.
input_dim = 6  # input_dimension = ['Open', 'High', 'Low', 'Close', 'Volume', 'vaderSentiment']

trainX = []
trainY = []
testX = []
testY = []

for i in range(seq_len, n_train-pred_days +1):
    trainX.append(train_data_scaled[i - seq_len:i, 0:train_data_scaled.shape[1]])
    trainY.append(train_data_scaled[i + pred_days - 1:i + pred_days, 0])

for i in range(seq_len, len(test_data_scaled)-pred_days +1):
    testX.append(test_data_scaled[i - seq_len:i, 0:test_data_scaled.shape[1]])
    testY.append(test_data_scaled[i + pred_days - 1:i + pred_days, 0])

trainX, trainY = np.array(trainX), np.array(trainY)
testX, testY = np.array(testX), np.array(testY)

# LSTM model
model = Sequential()
model.add(LSTM(64, input_shape=(trainX.shape[1], trainX.shape[2]), # (seq length, input dimension)
               return_sequences=True))
model.add(LSTM(32, return_sequences=False))
model.add(Dense(trainY.shape[1]))

model.summary()

# specify your learning rate
learning_rate = 0.01
# create an Adam optimizer with the specified learning rate
optimizer = Adam(learning_rate=learning_rate)
# compile your model using the custom optimizer
model.compile(optimizer=optimizer, loss='mse')

# Try to load weights
try:
    model.load_weights('./save_weights/lstm_weights.h5')
    print("Loaded model weights from disk")
except:
    print("No weights found, training model from scratch")
    # Fit the model
    history = model.fit(trainX, trainY, epochs=30, batch_size=32,
                    validation_split=0.1, verbose=1)
    # Save model weights after training
    model.save_weights('./save_weights/lstm_weights.h5')

    plt.plot(history.history['loss'], label='Training loss')
    plt.plot(history.history['val_loss'], label='Validation loss')
    plt.legend()
    plt.show()

# prediction
prediction = model.predict(testX)
print(prediction.shape, testY.shape)

# generate array filled with means for prediction
mean_values_pred = np.repeat(scaler.mean_[np.newaxis, :], prediction.shape[0], axis=0)

# substitute predictions into the first column
mean_values_pred[:, 0] = np.squeeze(prediction)

# inverse transform
y_pred = scaler.inverse_transform(mean_values_pred)[:,0]
print(y_pred.shape)

# generate array filled with means for testY
mean_values_testY = np.repeat(scaler.mean_[np.newaxis, :], testY.shape[0], axis=0)

# substitute testY into the first column
mean_values_testY[:, 0] = np.squeeze(testY)

# inverse transform
testY_original = scaler.inverse_transform(mean_values_testY)[:,0]
print(testY_original.shape)

# plotting
plt.figure(figsize=(14, 5))

# plot original 'Open' prices
plt.plot(dates, original_open, color='green', label='Original Open Price')

# plot actual vs predicted
plt.plot(test_dates[seq_len:], testY_original, color='blue', label='Actual Open Price')
plt.plot(test_dates[seq_len:], y_pred, color='red', linestyle='--', label='Predicted Open Price')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.title('Original, Actual and Predicted Open Price')
plt.legend()
plt.show()

# Calculate the start and end indices for the zoomed plot
zoom_start = len(test_dates) - 50
zoom_end = len(test_dates)

# Create the zoomed plot
plt.figure(figsize=(14, 5))

# Adjust the start index for the testY_original and y_pred arrays
adjusted_start = zoom_start - seq_len

plt.plot(test_dates[zoom_start:zoom_end],
         testY_original[adjusted_start:zoom_end - zoom_start + adjusted_start],
         color='blue',
         label='Actual Open Price')

plt.plot(test_dates[zoom_start:zoom_end],
         y_pred[adjusted_start:zoom_end - zoom_start + adjusted_start ],
         color='red',
         linestyle='--',
         label='Predicted Open Price')

plt.xlabel('Date')
plt.ylabel('Open Price')
plt.title('Zoomed In Actual vs Predicted Open Price')
plt.legend()
plt.show()
```

```python
#from tensorflow.keras.callbacks import EarlyStopping

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_aapl_1[features_1])
#scaled_data = scaler.fit_transform(df_aapl_2[features_2])
#scaled_data = scaler.fit_transform(df_aapl_3[features_3])

# 데이터를 시퀀스로 변환하는 함수
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data)-seq_length):
        x = data[i:(i+seq_length)]
        y = data[i+seq_length, 3]  # 'Close' 가격을 예측 대상으로 설정
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 14  # 시퀀스 길이 설정
X, y = create_sequences(scaled_data, seq_length)

# 데이터 분할 (시간 기반 분할)
train_size = int(len(X) * 0.8)  # 훈련 데이터 크기를 전체 데이터의 70%로 설정
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# 선택적으로 검증 세트 생성
val_size = int(len(X_train) * 0.2)  # 훈련 데이터의 20%를 검증 데이터로 설정
X_val, y_val = X_train[-val_size:], y_train[-val_size:]
X_train, y_train = X_train[:-val_size], y_train[:-val_size]

# LSTM 모델 구축 (LSTM 기본 활성화 함수 = tanh, sigmoid LSTM 그림 참조)
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25, activation='relu')) # 첫 번째 Dense 레이어에 ReLU 활성화 함수 적용
model.add(Dense(1))

# 모델 컴파일 (adam 학습률 = 0.001)
model.compile(optimizer='adam', loss='mean_squared_error')

# 조기 종료 콜백 설정
#monitor: 성능을 모니터링할 지표입니다. 여기서는 검증 데이터의 손실(val_loss)을 모니터링합니다.
#patience: 성능 개선이 없는 에폭을 얼마나 허용할지 지정합니다. 여기서는 10로 설정하여, 검증 손실이 10 에폭 동안 개선되지 않으면 학습을 중단합니다.
#restore_best_weights: True로 설정하면, 조기 종료 시점에서 가장 좋은 모델의 가중치를 복원합니다. 즉, 학습 과정에서 가장 낮은 검증 손실을 가진 모델의 가중치를 사용합니다.
#early_stopping = EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True)

# 모델 학습
#model.fit(X_train, y_train, batch_size=32, epochs=200, validation_data=(X_test, y_test), callbacks=[early_stopping])
model.fit(X_train, y_train, batch_size=32, epochs=300, validation_data=(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

# RMSE 계산 함수
def calculate_rmse(y_true, y_pred):
    return sqrt(mean_squared_error(y_true, y_pred))

# 모델 평가
predictions = model.predict(X_test)

# 평가 지표 계산
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
rmse = calculate_rmse(y_test, predictions)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```
